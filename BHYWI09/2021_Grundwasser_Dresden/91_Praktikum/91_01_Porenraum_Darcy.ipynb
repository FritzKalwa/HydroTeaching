### <a id='2_1'></a> 2.1 Porenraum

Der unterirdische Raum wird als ein poröses Medium betrachet.
Hohlräume in Gesteinen können von unterschiedlicher Natur sein, wobei Klüfte, Risse oder Poren je nach Art des Gesteins auftreten.
Damit haben die Hohlräume einen maßgeblichen Einfluss auf das Speichervermögen der Gesteinsformation sowie das Strömungsverhalten von Wasser.

<img src="img/poren_lockergestein.png" align='center' width="300">
<figcaption> Abb.3: Hohlräume in porösem Medium</figcaption>
</div>

Die Porosität $n$ gibt an, wie groß der Anteil des Hohlraumvolumens in Bezug auf das Gesamtvolumen der Gesteinsprobe, d. h. die Porosität ist der Quotient des Hohlraum- oder Porenvolumens $V_{p}$ vom Gesamtvolumen $V_{ges}$ des Gesteins.

\begin{equation}
n_{s} = \frac{V_{p}} {V_{ges}}
\label{1}
\end{equation}

mit:

* $n_s$... Porenanteil (Index s steht für soil) [-]
* $V_{ges}$... Gesamtvolumen $[m^{3}$]
* $V_{p}$... Porenvolumen $[m^{3}$]

Die Porosität ist u. a. abhängig von: 
* Form und Größe der Bodenteilchen,
* Lagerungs- bzw. Packungungsdichte.

Der Porenraum kann mit diversen Fluiden wie Luft oder Wasser gefüllt sein. 
Füllt Wasser diese Hohlräume völlig aus, so spricht man von gesättgten Verhältnissen. 
Sind die Hohlräume nur zum Teil mit Wasser gefüllt, so spricht man von ungesättigten Verhältnissen.

Geht man davon aus, dass der Porenraum nur mit Wasser sowie Luft gefüllt ist, gilt für das Volumen des Porenraums:
    
\begin{equation}
V_{p} = V_L + V_W
\label{2}
\end{equation}

mit:

* $V_{L}$... Volumen der Bodenluftphase $[m^{3}$]
* $V_{W}$... Volumen der Bodenwasserphase $[m^{3}$]

Die Porosität kann mit Hilfer der Korn- und Lagerungsdichte bestimmt werden.

Die Korndichte ist definiert als das Verhältnis der Trockenmasse $m_d$ zum Volumen der Körner $V_{s}$:
    
\begin{equation}
\rho_s = \frac{m_d}{V_s}
\label{3}
\end{equation}

mit:
    
* $\rho_s$... Korndichte $[\frac{g}{cm^3}$]
* $m_{d}$...  Trockenmasse $[g$]
* $V_s$...    Volumen der Körner $[cm^3$]

Die Lagerungsdichte entspricht dem Verhältnis der Trockenmasse $m_d$ zum Gesamtvolumen $V_{ges}$:
    
\begin{equation}
\rho_d = \frac{m_d}{V_{ges}}
\label{4}
\end{equation}

mit:
    
* $\rho_d$... Lagerungsdichte $[\frac{g}{cm^3}$]
* $m_{d}$...  Trockenmasse $[g$]
* $V_{ges}$...    Gesamtvolumen $[cm^3$]

Für die Berechnung der Porosität gilt folgender Zusammenhang:
    
\begin{equation}
n = 1 - \frac{\rho_d}{\rho_s}
\label{5}
\end{equation}

Da:
    
\begin{equation}
\begin{split}
n &= \frac{V_p}{V_{ges}} \\\ \\
&= \frac{V}{V} - \frac{V_{s}}{V} \\\ \\
&= 1 - \frac{\frac{m_d}{\rho_s}}{\frac{m_d}{\rho_d}} \\\ \\
&= 1 - \frac{\rho_d}{\rho_s} \\
\end{split}
\label{6}
\end{equation}

Für den teil- oder vollgesättigten Untergrund wird das Wasser weiter in mobiles und immobiles Wasser differenziert.
Dabei wird das immobile Wasser durch z. B. Kapillar- und Bindungskräfte so stark an die Feststoffmatrix fixiert, dass es nicht beweglich und damit immobil ist.
Je kleiner die Porenräume sind, umso größer ist der Einfluss von Kapillarkräften, so dass bei Lockersedimenten in Abhängigkeit von der Korngröße ein großer Teil des Porenraums mit immobilien Wasser gefüllt sein kann.
Die verbleibende mobile Porosität, häufig auch als effektive Porosität bezeichnet, kennzeichnet den Anteil der Gesamtporosität, der von mobilem Wasser okkupiert ist.

Es ergibt sich folgender Zusammenhang:

\begin{equation}
n = n_{mob} + n_{im}
\label{7}
\end{equation}

Wie groß der Anteil der mobilen Porosität ist hängt von der Bodenart ab.

### <a id='2_2'></a> 2.2 Wasserbewegung

Das Darcy-Gesetz beschreibt den linearen Zusammenhang zwischen hydraulischem Gradienten und Strömungsgeschwindigkeit: 

\begin{equation}
q = -K \frac{\Delta{h}} {\Delta{l}}
\end{equation}
Hierbeit stellt $q$ die Strömungsgeschwindigkeit dar, $\Delta{h}$ repräsentiert den Druckhöhenunterschied zwischen den beiden betrachteten Punkten ($\Delta{h} = h_2 - h_1$) und $\Delta{l}$ die Fließweglänge (Abstand zwischen den beiden betrachteten Punkten).
Der Proportionalitätsfaktor $K$ beschreibt die hydraulische Leitfähigkeit (auch $K$-Wert oder vor allem in der deutschsprachigen Literatur $k_f$-Wert genannt).
Dies stellt den interessantesten Parameter eines porösen Mediums dar, den wir üblicherweise in einem Säulenexperiment (vgl. Darcy-Experiment) wie diesem bestimmen wollen - doch dazu gleich mehr.
Während sich der Druckhöhenunterschied an Anfang und Ende der Säule relativ einfach über die Messung der Wasserspiegel an den beiden Stellen erfassen lässt und auch die Längenmessung der Säule keine große Herausforderung darstellt, ist die Messung der Strömungsgeschwindigkeit nicht auf einfache Weise direkt möglich.
Wir errechnen Sie uns deshalb indirekt über den gemessenen Volumenstrom $Q$, den wir einfach zur Querschnittsfläche der Säule $A$ ins Verhältnis setzen, und somit nur noch auslitern müssen, wie viel Wasservolumen pro Sekunde/Minute/o.Ä. aus der Säule herausläuft:

\begin{equation}
q= \frac{Q} {A}
\end{equation}

Der $K-Wert$ kann um mehrere Größenordnungen schwanken, je nachdem wie groß die Porenräume sind, die durchströmt werden.
Üblicherweise wird der Wert, da er eine Leitfähigkeit (analog zur elektrischen Leitfähigkeit) darstellt, in gröberen Sedimenten (z.B. Sand oder Kies) größer, da die größeren Poren hier deutlich weniger Fließwiderstand bieten als die kleineren Poren in feinen Sedimenten (z.B. Schluff oder Ton).

<div class="alert alert-block alert-info">
Mit dem folgenden Slider können Sie sehen, wie sich ihr Durchfluss unter verschiedenen Bedingungen ändert. Der ausgegebene Wert entspricht dem Durchfluss Q [$m^3$/s]. Die Startwerte sind so gewählt, dass die den möglichen Einstellungen des Experimentes entsprechen.

```python
from IPython.display import display
import ipywidgets as widgets
import matplotlib.pyplot as plt
from matplotlib import patches
from ipywidgets import interact, interactive, fixed, interact_manual
import numpy as np

def Q(K,A,h,l):
    try:
        print("spezifischer Durchfluss q = %8.2e"% (K * (h/l)),       " m/s")
        print("spezifischer Durchfluss q = %8.4f"% (K * (h/l)*86400), " m/d")
        print("             Durchfluss Q = %8.3f"% (K * A * (h/l)*1000), "l/s")
        print("             Durchfluss Q = %8.3f"% (K * A * (h/l)*1000*60), "l/min")
        print("             Durchfluss Q = %8.3f"% (K * A * (h/l)*1000*60*60), "l/h")

        r = np.sqrt(A/np.pi)
        fig = plt.figure(figsize=(9,6))
        ax = fig.add_subplot(1, 1, 1, aspect='equal')

        ax.hlines(0, l, 1.5*l, color = 'blue', linewidth = 10000*K * (h/l))
        plt.text(l+r, 0.5 * r, 'Q', ha='center', color = 'black')

        e1 = patches.Ellipse((0, 0), r, r*2,
                        linewidth=2, fill=False, zorder=2)
        e2 = patches.Ellipse((l, 0), r, r*2,
                        linewidth=2, fill=False, zorder=2)
        ax.add_patch(e1)
        ax.add_patch(e2)
        ax.hlines(r, 0, l, color = 'black')
        ax.hlines(-r, 0, l, color = 'black')
        ax.hlines(0, 0, -0.5*l, color = 'blue')
        ax.vlines(-0.5*l, 0, h, color = 'blue')
        ax.vlines(-0.5*l-r*0.375, 0, h, color = 'red')
        ax.hlines(0, -0.5*l-r*0.25, -0.5*l-r*0.5, color = 'red')
        ax.hlines(h, -0.5*l-r*0.25, -0.5*l-r*0.5, color = 'red')
        plt.text(-0.9*l, h*0.5,  ' Delta h', ha='center', color = 'red')

        ax.vlines(0, r*1.25, r*1.5, color = 'red')
        ax.vlines(l, r*1.25, r*1.5, color = 'red')
        ax.hlines(r*1.375, 0, l, color = 'red')
        plt.text(l*0.5, r*1.6,  ' Laenge l', ha='center', color = 'red')

        ax.vlines(-l,0,0.0000001, color = 'white')

        ax.axes.xaxis.set_visible(False)
        ax.axes.yaxis.set_visible(False)
        
    except:
        print('Division durch 0. Passen Sie Ihre Werte an.')

w = interactive(Q,K=widgets.FloatLogSlider(min=-8, max=-2,value=0.0005, step=0.0001,description='K [m/s]:',readout_format='8.2e',),
                        A=widgets.FloatLogSlider(min=-4, max=-2, value=0.0050, step=0.001,description='A [$m^2$]:'),
                        h=widgets.FloatLogSlider(min=-2, max=-0.2, value=0.2, step=0.001,description='$\Delta$h [m]:'),
                        l=widgets.FloatLogSlider(min=-2, max=-0.3, value=0.1, step=0.001,description='$\Delta$l [m]:'),
                       )
display(w)
```

# <a id='3'></a> 3 Porenraumbestimmung


### 3.1 <a id='3_1'></a> Benötigte Materialien

Um den Porenraum von Lockersedimenten zu charakterisieren, benötigen Sie:
* zwei Kaffeepötte (oder ähnliche Gefäße),
* eine einfache Küchenwaage,
* Wasser,
* Lockersedimente (z. B. Sande mit unterschiedlicher Korngröße, Kies).

Wir benötigen ca. 'einen halben Liter' der Sedimente, um das Gefäß zu füllen.
Diese Sedimente sollten Sie in ihrem näheren Umfeld finden können (Spielplatz, Baumarkt).
Verwenden Sie möglichst trockene Sedimente (wir gehen davon aus, dass die Sedimente komplett trocken sind).
Trocknen Sie die Sedimente ggf. im Vorfeld (Sonne, Ofen bei maximal $60^{\circ}C).

### 3.2 <a id='3_2'></a> Experimentelle Bestimmung des Porenraums

Mit den folgenden Anweisungen können Sie den Porenraum (totale und mobile Porosität) Ihrer Sedimente charakterisieren.

* Notieren Sie Ihre Erwartungswerte für die totale und mobile Porosität,
* Bestimmen Sie das Volumen des Gefäßes $V_t$, in welches Sie die Sedimente für den Versuch füllen möchten.
Ermitteln Sie das Volumen über die Wassermenge (bzw. das Gewicht des Wassers), welches das Gefäß bei kompletter Füllung fasst.
* Leeren Sie das Gefäß, trocknen Sie dieses ab und füllen Sie anschliessend das Gefäß komplett mit Sediment.
* Erfassen Sie die Wassermenge $V_1$, welche in das komplett mit Sediment gefüllte Gefäß passt (z. B. mit Hilfe der gravimetrisch ermittelten Differenz der Massen).
* Erfassen Sie die Wassermenge $V_2$, welche aus dem komplett mit Sediment gefüllte Gefäß wieder ausfliesst (z. B. mit Hilfe der gravimetrisch ermittelten Differenz der Massen).

* Mit Hilfe der folgenden Gleichungen können Sie nun die totale und entwässerbare (dränbare) Porosität $S_y$ berechnen:

\begin{equation}
n_{t} = \frac{V_1} {V_t}
\end{equation}  


\begin{equation}
S_y = \frac{V_2} {V_t}
\end{equation}

In diesem Experiment soll angenommen werden, dass die dränbare Porosität ($S_y$) in etwa der mobilen Porosität entspricht. Da dann gilt

\begin{equation}
S_y \overset{!}{=} n_{mob}
\end{equation}

gilt auch:
    
\begin{equation}
n_{mob} = \frac{V_2} {V_t}
\end{equation}

* Vergleichen Sie die experimentell ermittelten Porositäten mit Ihren Erwartungswerten und analysieren Sie ggf. Unsicherheiten und Fehlerquellen.

Für die Berechnung können Sie auch nachfolgendes Skript verwenden.

```python
from IPython.display import display
import ipywidgets as widgets
import matplotlib.pyplot as plt

print('Berechnung der totalen (n_t) und mobilen (n_mob) Porosität:')

def n(V_1,V_2,V_t):
    n_t = V_1/V_t
    n_mobil = V_2/V_t
    print("  n_t = %1.2f"% n_t)
    print("n_mob = %1.2f"% n_mobil)
    
    try:
        plt.style.use('seaborn-bright')
        labels = 'Nutzporenraum', 'Haftwasserraum'
        sizes = [n_mobil/n_t, 1-(n_mobil/n_t)]
        colors = ['lightblue', 'green']
    

        plt.pie(sizes,
                labels=labels,
                colors=colors,
                autopct='%1.2f%%',
                startangle=90)

        plt.axis('equal')

        plt.show()
    except:
        print('Division durch 0. Passen Sie ihre Werte an')
    
    

w = interactive(n,      V_1=widgets.BoundedFloatText(
                        value=0.3,
                        min=0,
                        max=10000000000.0,
                        step=0.001,
                        description='V_1 [dm^3]',
                        ),
                        V_2=widgets.BoundedFloatText(
                        value=0.25,
                        min=0,
                        max=10000000000.0,
                        step=0.001,
                        description='V_2 [dm^3]',
                        ),
                        V_t=widgets.BoundedFloatText(
                        value=1,
                        min=0.001,
                        max=1000.0,
                        step=0.001,
                        description='V_t [dm^3]',
                        )
                       )
display(w)
```

## <a id='4'></a> 4 Aufbau des Darcy-Experimentes

### <a id='4_1'></a> 4.1 Benötigte Materialien 
Zunächst benötigen Sie ein Sediment, das Sie untersuchen können.
Versuchen Sie, in Ihrer Umgebung eine kleine Menge (ca. 300-500 ml) sandigen Sedimentes zu finden.
Achten Sie darauf, dass sich in Ihrem Sediment möglichst wenig größere Steine befinden.
Außerdem sollten sich auch möglichst wenig feinere Partikel (Ton, Organik, o.Ä.) im Sediment befinden.
Alles, was zu sehr "matscht" scheidet also aus. Wenn Sie kein "natürliches" Sediment finden, können Sie im Zweifelsfall auch eine Probe aus einem Sandkasten für Kinder nehmen.
**Bitte in dem Fall darauf achten, dass Sie wirklich nicht mehr als die benötigten 0,3-0,5 l entnehmen!**
<br>Reiben Sie Ihre Sedimentprobe zu Beginn zwischen den Fingern und versuchen Sie grob festzustellen, welche Sedimentart dominiert.
Als Faustregel können Sie sagen, dass alles, was größer ist als ein Streichholzkopf als "Kies" bezeichnet wird.
Alles zwischen Streichholzkopfgröße und Weizengrieß bezeichnet man als Grobsand und alles was eher an Mehl erinnert, bezeichnet man als Feinsand.
Notieren Sie Ihre Einschätzung und schauen Sie in Abbildung 2 nach, welchen K-Wert Sie erwarten würden.
<br>Um den K-Wert zu bestimmen, wird eigentlich eine Säule mit gleichbleibendem Fließquerschnitt benötigt, die sich an beiden Enden wasserdicht abschließen lässt.
In unserem Fall versuchen wir es etwas anders und wählen eine handelsübliche PET-Flasche, die idealerweise die Form eines Zylinders aufweist, also ohne Verengungen oder Auswölbungen gestaltet ist.
Aus dieser wird der Boden herausgeschnitten und der Deckel perforiert.
Danach wird die "Säule" mit Sediment befüllt und kopfüber auf einem Ständer in eine Schüssel gestellt, sodass Wasser, das durch die aufgeschnittene Öffnung hinzugegeben wird durch die Säule in die Schüssel fließen kann.
Hier eine Liste der benötigten Gegenstände (sie können aber auch kreativ davon abweichen): 

* eine PET-Flasche (idealerweise Einweg)
* eine etwa 5-8 cm hohe Schüssel mit ebenem Boden
* ein Bandmaß oder Zollstock
* ein Messer (idealerweise ein Cuttermesser)
* ein Smartphone oder eine Stoppuhr
* ein wasserdurchlässiges Netzgewebe (z.B. ein Kaffee- oder Teefilter)
* eine Küchenwaage
* einen Permanentmarker
* eine Hand voll Kies (alternativ auch getrocknete Erbsen oder Weizenkörner)
* das Sediment, das sie untersuchen wollen - Die Methode ist besser für sandige Sedimente geeignet, also alles zwischen 
* ein "Sieb" zur gleichmäßigen Verteilung des zugegebenen Wassers auf der Sedimentoberfläche (z.B. ein löchriger Joghurtbecher)
* ein (wasserdurchlässiger) Ständer für die umgedrehte Flasche, z.B. ein weiterer löchriger Becher, ein Tee-Stövchen, o.Ä.
* ein "Verdichter": Also einen Stab, o.Ä., der einen Durchmesser von 3-5 cm hat. Hier eignen sich z.B. die handelsüblichen Gewürzgläser oder auch die Rückseite eines Schraubenziehers.

### <a id='4_3'></a> 4.3 Einbau der Sedimente 
Setzen Sie nun den abgesägten Flaschenboden verkehrt herum in die umgedrehte Flasche ein, drücken Sie ihn bis ganz nach unten und füllen Sie eine 2-4 cm mächtige Kiesschicht ein.
Nun gilt es, das zu untersuchende Sediment einzufüllen, ohne dass es sich mit der Kiesschicht vermischt.
Dafür müssen Sie (je nach Flaschen- und Filtergröße) ein bisschen improvisieren: Sie können z.B. ihren Kaffe-/Teefilter so zurechtschneiden, dass er die beiden Schichten sicher trennt oder einen Teil der Sedimentschicht in den Filter eingeben, diesen fest auf die Kiesschicht pressen und danach das restliche Sediment einfüllen.
Dazu feuchten Sie das Sediment leicht an (bei 300 ml Sediment ca. 50 ml Wasser), füllen zunächst eine Schicht von 2-3 cm ein. Diese verdichten Sie im Anschluss, indem Sie die Sedimentoberfläche mit dem Verdichter mehrmals abklopfen.
Wie viel Kraft Sie dabei einsetzen hängt von der Stabilität der Konstruktion ab.
Grundsätzlich gilt: Je mehr, desto realitätsnäher.

Messen und notieren Sie die Länge der finalen Sedimentsäule (**ohne** Kiesschicht!).
Nun haben Sie ihr $\Delta{l}$.

## <a id='5'></a> 5 Darcy-Experiment: Durchführung


### <a id='5_1'></a> 5.1 Vorbereitung des Experimentes

Das Experiment sollten Sie in einer Badewanne, der Dusche, dem Waschbecken oder draußen im Freien durchführen, um sich die Arbeit mit dem Feudel zu sparen.

* Zunächst stellen Sie die Schüssel auf dem Boden ab und setzen den "Ständer" mit der umgedrehten und mit Sand befüllten Flasche in die Mitte der Schüssel.

Das Darcy-Gesetz ist nur unter gesättigten Bedingungen gültig.
Es gilt also, die zu untersuchende Sedimentschicht möglichst gut aufzusättigen.

* Befüllen Sie die Flasche deshalb stetig mit Wasser, sodass die "Säule" kontinuierlich durchströmt wird.
Versuchen Sie, das Sediment mit mindestens dem 5-fachen des "Säulenvolumens" zu durchströmen, besser mit dem 10-fachen (also bei 300 ml Sediment: 1,5-3 l Wasser).
Kochen Sie das Wasser vorher ab und füllen Sie es in eine Flasche. Dadurch strippen Sie gelöste Gase im Wasser aus.

Die Schüssel sollte nach einiger Zeit bis zum Rand mit Wasser gefüllt sein und überlaufen.
Falls die Schüssel so groß sein sollte, dass Sie sie auch nach Durchspülung mit dem 10-fachen Probenvolumen nicht bis zum Rand füllen konnten, können Sie das fehlende Wasser auch direkt in die Schüssel geben.

* Nehmen Sie ihr Maßband und messen Sie entlang der Flasche eine Distanz von 20 cm ab, beginnend vom Wasserstand der Schüssel bis hin zur abgesägten Öffnung.
Falls Ihre Flasche deutlich kürzer/länger ist, können Sie auch 15, bzw. 30 cm vom Wasserspiegel abmessen. Markieren Sie mit dem Permanentmarker die Position mit einem Strich.
* Machen Sie einen weiteren Strich jeweiles 2 cm oberhalb und unterhalb des ersten Striches.

Wenn Sie nun kontinuierlich Wasser nachgeben und darauf achten, dass der Wasserspiegel sich immer mehr oder weniger an der Position des Striches befindet, stellen Sie einen Druckhöhenunterschied (=$\Delta{h}$) von 0,2 m (bzw. 0,15/0,3 m) sicher.
Wenn Sie nun diesen Druckhöhenunterschied durch die Länge Ihrer Sedimentsäule ($\Delta{l}$) teilen (siehe Kapitel 3.3), erhalten Sie den hydraulischen Gradienten $\Delta{h}$/$\Delta{l}$.

### <a id='5_2'></a> 5.2 Durchführung des Experimentes 
Wiegen Sie die mit Wasser gefüllte Flasche mit der Küchenwaage und notieren Sie sich das Gewicht. 
Befüllen Sie mit dieser Flasche nun in regelmäßigen Abständen Ihre Säule, sodass der Wasserstand leicht über der Markierung liegt.
Halten Sie die Stoppuhr bereit, denn sobald der sinkende Wasserpegel die Markierung erreicht, muss die Zeit gemessen werden.
Nun gilt es, den Wasserstand im Mittel bei ungefähr 20 cm zu halten.
Dazu füllen Sie Ihre Säule bis zum Strich 2 cm oberhalb der 20 cm-Markierung mit Wasser, warten bis der Wasserstand zum Stricht 2 cm unterhalb der 20 cm-Markierung abgefallen ist und füllen dann wieder bis zum oberen Strich auf.
Je nach Sedimentzusammensetzung kann das bedeuten, dass Sie alle 5-30 Sekunden frisches Wasser zugeben müssen.
<br>
Es ist es Ihnen überlassen, wie lange Sie den Versuch durchführen.
Es wäre aber auch hier empfehlenswert, mindestens so lange abzuwarten, dass das 3-fache des Sedimentprobenvolumens durch die Säule geflossen ist.
Nachdem Sie sich dazu entschieden haben den Versuch zu stoppen, notieren Sie sich die verlaufene Zeit und wiegen die PET-Flasche erneut.
Wenn Sie jetzt die Gewichtsdifferenz der Flasche durch die gemessene Zeit teilen, können Sie den Durchfluss errechnen (bedenken Sie: **1 ml = 1 g Wasser** )

Führen Sie diesen Versuch min. drei Mal durch und mitteln Sie den gemessenen Durchfluss ($Q$).
<br> **Nun haben Sie alle Parameter erhoben, die Sie zur Bestimmung der hydraulischen Leitfähigkeit benötigen**

Für die Berechnung können Sie das nachfolgende Script verwenden.

```python
from IPython.display import display
import ipywidgets as widgets
import matplotlib.pyplot as plt
from matplotlib import patches
from ipywidgets import interact, interactive, fixed, interact_manual
import numpy as np

def K(Q,A,h,l):
    print("hydraulische Leitfähigkeit K = %8.2e"% (Q/(A*(h/l)))," m/s")

w = interactive(K,      Q=widgets.BoundedFloatText(
                        value=0.00001,
                        min=0,
                        max=100000000.0,
                        step=0.000001,
                        description='Q [m^3/s]',
                        ),
                        A=widgets.BoundedFloatText(
                        value=1,
                        min=0.0000000000001,
                        max=10000000.0,
                        step=0.00001,
                        description='A [m^2]',
                        ),
                        h=widgets.BoundedFloatText(
                        value=0.00000001,
                        min=0.001,
                        max=10000000.0,
                        step=0.00001,
                        description='h [m]',
                        ),
                        l=widgets.BoundedFloatText(
                        value=1,
                        min=0.0000000001,
                        max=100000000.0,
                        step=0.00001,
                        description='$\Delta$l [m]:'
                        )
                       )
display(w)
```
