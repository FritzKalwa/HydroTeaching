# Decay

This notebook demonstrate radioactive decay for a decay chain: A -> B -> B

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

Define the time span *t* and set the arrays for time and concentration

```python
timespan = 20
t = np.arange(0,timespan,1)
C_A = [0 for x in range(timespan)]
C_B = [0 for x in range(timespan)]
C_C = [0 for x in range(timespan)]
C_sum = [0 for x in range(timespan)]
```

Define the initial concentrations for *A*, *B*, and *C*, and the decay rates for A and B

```python
C_A[0]=1000
C_B[0]=0
C_C[0]=0
C_sum[0]=C_A[0]+C_B[0]+C_C[0]
lambda_A = 0.5
lambda_B = 0.2
```

Compute the concentration, skip the computation for the initial time

```python
for x in t:
    if x > 0:
        C_A[x]=C_A[x-1]-C_A[x-1]*lambda_A
        C_B[x]=C_B[x-1]+C_A[x-1]*lambda_A-C_B[x-1]*lambda_B
        C_C[x]=C_C[x-1]+C_B[x-1]*lambda_B
        C_sum[x]=C_A[x]+C_B[x]+C_C[x]
```

Plot the results

```python
    fig, ax = plt.subplots()
    ax.plot(t,C_A)
    ax.plot(t,C_B)
    ax.plot(t,C_C)
    
    ax.set(xlabel='time', ylabel='conc.',
           title='Decay')
    ax.grid()
    plt.legend(('A', 'B', 'C'))
    
    plt.show()
```
